Generate random data for exponential, uniform and normal distribution

```python
exp_data = np.random.exponential(scale=1.0, size=10000)
```

```python
uniform_data = np.random.rand(10000)
```

```python
normal_data = np.random.randn(10000)
```

Create histograms for exponential distributions

```python
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
```

```python
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
```

```python
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.title("Exponential Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Create histograms for Uniform distributions

```python
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
```

```python
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.legend(loc="upper left")
plt.title("Uniform Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Create histograms for normal distributions

```python
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
```

```python
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.legend(loc="upper left")
plt.title("Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(normal_data, density=True, bins=50, histtype="step", color="purple", label="Normal")
plt.legend(loc="upper left")
plt.title("Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

plot the Z = X**2 + Y**2 in a 3D plot

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2
```

Create the 3D plot


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title("3D Plot of Z = X + Y")
plt.show()
```

```python
data = sns.load_dataset("tips")
```

Extract only numeric values for calculate pearson correlation

```python
numeric_data = data.select_dtypes(include=[np.number])
```

```python
print(numeric_data.head())
```

```python
pearson_corr = numeric_data.corr(method='pearson')
```

```python
print(pearson_corr)
```

Show the Pearson Correlation in the Heatmap

```python
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', linewidths=1, linecolor='black')
```

```python
plt.title("Pearson Correlation Matrix")
plt.show()
```

```python
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', linewidths=1, linecolor='black')
plt.title("Pearson Correlation Matrix")
plt.show()
```

```python
spearman_corr = numeric_data.corr(method='spearman')
```

```python
print(spearman_corr)
```

Show the spearman Correlation in the Heatmap

```python
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', linewidths=1, linecolor='black')
```
